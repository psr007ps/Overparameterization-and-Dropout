```
import keras
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Activation
from keras.layers import Conv2D, MaxPooling2D
from keras import backend as K
```

```
batch_size = 128
num_classes = 10
epochs = 12
```

```
# the data, split between train and test sets
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```
x_train = x_train.reshape(60000, 784)
x_test = x_test.reshape(10000, 784)
```

```
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
x_train /= 255
x_test /= 255
```

```

#take 10000 samples out of 50000 for training
from sklearn.model_selection import train_test_split
x_train, X_test, y_train, Y_test = train_test_split(x_train, y_train, stratify=y_train, test_size=5/6)
```

```
from scipy import stats
x_train = stats.zscore(x_train, axis=1)
x_test = stats.zscore(x_test, axis=1)
```

```
#for question 3
y = y_train
y_t = y_test
x = x_train
x_t = x_test
```

```
# convert class vectors to binary class matrices
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)
```

**Question 1:**

Neural Network Setup

```
from keras.callbacks import CSVLogger
def neural_net(question, x_train, y_train, x_test, y_test, K, P, epochs=80, batch_size=10):
  for p in P:
    for k in K:
      #Model
      model = Sequential()
      #input layer
      model.add(Dense(k, input_shape=(784,), kernel_initializer='he_normal'))
      model.add(Activation('relu'))
      model.add(Dropout(p))
      #First Hidden Layer
      model.add(Dense(k))
      model.add(Activation('relu'))
      model.add(Dropout(p))
      #Output Layer
      model.add(Dense(10))
      model.add(Activation('softmax'))

      model.compile(loss=keras.losses.categorical_crossentropy,
                    optimizer=keras.optimizers.Adam(),
                    metrics=['accuracy'])
      
      csv_logger = CSVLogger(question+".log", append=True)

      model.fit(x_train, y_train,
                batch_size=batch_size,
                epochs=epochs,
                verbose=1,
                validation_data=(x_test, y_test),
                callbacks=[csv_logger])
      
      score = model.evaluate(x_test, y_test, verbose=0)
      print('Test loss:', score[0])
      print('Test accuracy:', score[1])
```

**Question 2.1:** 

---

---



```
batch_size = 10
K = [1, 3, 5, 10, 15, 20, 30, 40] 
neural_net("q2.1", x_train, y_train, x_test, y_test, epochs=80, batch_size=10, K=K, P=[1])
```
